cv2.resize().

```python
import cv2

capImg = cv2.VideoCapture('DorZn.avi')

new_width = 640
new_height = 480

while capImg.isOpened():
    # Получаем кадр из видеопотока файла
    ret, frame = capImg.read()

    # Если кадры закончились
    if frame is None:
        break

    resized_frame = cv2.resize(frame, (new_width, new_height))

    cv2.imshow("Resized Video", resized_frame)

    flipped_frame = cv2.flip(frame, 1)

    cv2.imshow("Flipped Video", flipped_frame)

    # Организуем выход из цикла по нажатию клавиши, ждем 30 миллисекунд нажатия, записываем код нажатой клавиши
    key_press = cv2.waitKey(30)

    # Если код нажатой клавиши совпадает с кодом «q»,
    if key_press == ord('q'):
        break  # то прервать цикл while

# Освобождаем память от переменной capImg
capImg.release()

cv2.destroyAllWindows()
```

Задание 9

Подберите границы цветовых диапазонов для поиска объектов 
других цветов (красный, желтый, зеленый и т. д.) на видео.

```python
import numpy as np
import cv2

# Связываем видеопоток файла video.avi с переменной capImg
capImg = cv2.VideoCapture('DorZn.avi')

while capImg.isOpened():
    # Получаем кадр из видеопотока файла
    ret, frame = capImg.read()

    # Если кадры закончились, то прерываем цикл
    if frame is None:
        break

    # Переводим кадр в цветовое HSV-пространство
    frame_hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

    # Задаем нижние и верхние границы цветового фильтра
    low_Red = np.array([150, 100, 100], dtype="uint8")
    high_Red = np.array([180, 255, 255], dtype="uint8")

    low_Blue = np.array([110, 70, 100], dtype="uint8")
    high_Blue = np.array([130, 255, 255], dtype="uint8")

    low_YellowOrange = np.array([15, 50, 50], dtype="uint8")
    high_YellowOrange = np.array([60, 255, 255], dtype="uint8")

    # Применяем маску по цвету к фрагменту кадра
    red_mask = cv2.inRange(frame_hsv, low_Red, high_Red)
    blue_mask = cv2.inRange(frame_hsv, low_Blue, high_Blue)
    yelloworange_mask = cv2.inRange(frame_hsv, low_YellowOrange, high_YellowOrange)

    # #Показываем области кадров с объектами в окне video_mask
    cv2.imshow("Red Objects", red_mask)
    cv2.imshow("Blue Objects", blue_mask)
    cv2.imshow("Green Objects", yelloworange_mask)

    # В окне video_frame показываем вырезку из кадра
    cv2.imshow("Frame", frame)

    # Организуем выход из цикла по нажатию клавиши
    key_press = cv2.waitKey(0)

    # Если код нажатой клавиши совпадает с кодом «q»,
    if key_press == ord('q'):
        break

# Освобождаем память от переменной capImg
capImg.release()

cv2.destroyAllWindows()
```

Задание 10
1. Меняя в функции cv2.drawContours() параметры
contoursIdx, color, thickness (3, 4 и 5-й аргументы 
по счету соответственно), выведите на экран по отдельности 
внешний и внутренний контуры, измените цвет контура и толщину линий контура.
2. Подготовьте с помощью редактора Paint изображение с несколькими объектами, окрашивая их в различные цвета. Подбирая 
параметры цветового фильтра, выведите на экран поочередно 
все контуры каждого объекта.

```python
import numpy as np
import cv2

# Задаем путь к файлу с картинкой
fileName = 'Orang.jpg' 

# Считываем данные графического файла в переменную image
image = cv2.imread(fileName)

# Конвертируем исходное изображение в цветовую модель HSV и результат записываем в переменную hsv_img
hsv_img = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Подбираем параметры цветового фильтра для выделения нашего объекта
hsv_min = np.array((2, 28, 65), np.uint8)
hsv_max = np.array((26, 238, 255), np.uint8)

# Применяем цветовой фильтр к исходному изображению, результат записываем в переменную hsv_msk
hsv_msk = cv2.inRange(hsv_img, hsv_min, hsv_max) 

# Ищем контуры и записываем их в переменную contours
contours, hierarchy = cv2.findContours(hsv_msk, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)

# Отображаем внешние контуры поверх исходного изображения (цвет синий, толщина линии 3)
cv2.drawContours(image, contours, -1, (255, 0, 0), 3, cv2.LINE_AA, hierarchy, 2)

# Выводим итоговое изображение
cv2.imshow("External Contours", image)
cv2.waitKey(0)
cv2.destroyAllWindows()

# Создаем копию исходного изображения для вывода внутренних контуров
image_internal = image.copy()

# Выводим внутренние контуры
for i in range(len(contours)):
    cv2.drawContours(image_internal, contours, i, (0, 255, 0), 2, cv2.LINE_AA, hierarchy, 0)

# Выводим итоговое изображение с внутренними контурами
cv2.imshow("Internal Contours", image_internal)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

ЗАДАНИЕ 11
1. Используя этот же рисунок, выполните фильтрацию контуров 
объектов желтого цвета и выведите в окне контур самого большого объекта.
2. Применяя наложение суммы масок, выделите из рисунка контуры объектов различных цветов, например красных и желтых 
объектов с отсевом мелких контуров и т. д

```python
import numpy as np
import cv2

# Задаем путь к файлу с картинкой
fileName = 'Naturmort.jpg' 

# Считываем данные графического файла в переменную
image = cv2.imread(fileName)

# Конвертируем изображение в цветовую модель HSV
hsv_img = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Задаем параметры цветового фильтра для выделения желтого цвета
yellow_min = np.array([20, 100, 100], np.uint8)
yellow_max = np.array([30, 255, 255], np.uint8)

# Применяем цветовой фильтр к изображению
yellow_mask = cv2.inRange(hsv_img, yellow_min, yellow_max)

# Находим контуры объектов желтого цвета
yellow_contours, _ = cv2.findContours(yellow_mask, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)

# Находим самый большой контур желтого цвета и рисуем его
if len(yellow_contours) > 0:
    max_yellow_contour = max(yellow_contours, key=cv2.contourArea)
    cv2.drawContours(image, [max_yellow_contour], -1, (0, 255, 255), 3)
    cv2.imshow("Yellow Contour", image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
else:
    print("Желтые контуры не найдены.")

# Определение диапазонов цветов для других объектов (например, красного)
red_min = np.array([0, 100, 100], np.uint8)
red_max = np.array([10, 255, 255], np.uint8)

# Применение цветовых фильтров к изображению
red_mask = cv2.inRange(hsv_img, red_min, red_max)

# Наложение суммы масок (для выделения красных и желтых объектов)
combined_mask = yellow_mask + red_mask  

# Нахождение контуров на изображении с учетом суммы масок
combined_contours, _ = cv2.findContours(combined_mask, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)

# Отображение контуров объектов различных цветов
for contour in combined_contours:
    area = cv2.contourArea(contour)
    # Отсев мелких контуров (площадь < порога)
    if area > 1500:
        cv2.drawContours(image, [contour], -1, (0, 255, 0), 3)  # Цвет зеленый, толщина линии 3

# Вывод итогового изображения
cv2.imshow("Combined Contours", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Задание 12

В программе более верно подобрать значения цветов для определения дорожных знаков и обвести их контуром.

```python
import numpy as np
import cv2
import time

capImg = cv2.VideoCapture('DorZn.avi')

while(capImg.isOpened()):
    ret, frame = capImg.read()

    if frame is None:
        break

    frame_hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    crop_frame = frame[160:270, 150:650]
    crop_frame_hsv = frame_hsv[160:270, 150:650]

    blue_min = np.array([100, 100, 100], np.uint8)
    blue_max = np.array([120, 255, 255], np.uint8)

    yellow_min = np.array([20, 100, 100], np.uint8)
    yellow_max = np.array([30, 255, 255], np.uint8)

    red1_min = np.array([0, 100, 100], np.uint8)
    red1_max = np.array([5, 255, 255], np.uint8)

    red2_min = np.array([170, 100, 100], np.uint8)
    red2_max = np.array([180, 255, 255], np.uint8)

    blue_mask = cv2.inRange(frame_hsv, blue_min, blue_max)
    yellow_mask = cv2.inRange(frame_hsv, yellow_min, yellow_max)
    red1_mask = cv2.inRange(frame_hsv, red1_min, red1_max)
    red2_mask = cv2.inRange(frame_hsv, red2_min, red2_max)

    full_mask = blue_mask + yellow_mask + red1_mask + red2_mask

    # Находим контуры
    contours, hierarchy = cv2.findContours(full_mask.copy(), cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

    # Отсев мелких контуров и рисование прямоугольников
    for contour in contours:
        rect = cv2.minAreaRect(contour)
        area = int(rect[1][0] * rect[1][1])

        if area > 100:
            box = cv2.boxPoints(rect)
            box = np.int0(box)
            cv2.drawContours(frame, [box], 0, (0, 255, 255), 2)

    cv2.imshow("Mask", full_mask)
    cv2.imshow("Original Frame", frame)
    cv2.imshow("Cropped Frame", frame)

    # Задержка
    time.sleep(0.1)

    # Завершение по нажатию клавиши 'q'
    if cv2.waitKey(30) & 0xFF == ord('q'):
        break

capImg.release()
cv2.destroyAllWindows()
```
